```python
from bardapi import Bard
import os
```

```python
os.environ['_BARD_API_KEY']='KEY'
```

```python
def generate_message(aluno):
  mensagens = []
  for index, row in aluno.iterrows():
    aluno_nome = row['Nome']
    nota = row['Nota']
    input_text = f"Crie uma mensagem para o aluno {aluno_nome} sobre suas notas. Se a nota ({nota}) for maior que 7, o parabenize; caso contrário, dê conselhos para melhorar (máximo de 100 caracteres)."
    bard_output = Bard().get_answer(input_text)['content']
    mensagens.append(bard_output)
  return mensagens
```

```python
aluno_mensagens = generate_message(notas_df)
```

```python
aluno_mensagens
```

```python
for i, message in enumerate(aluno_mensagens):
    print(f"Aluno: {notas_df['Nome'][i]}, Mensagem: {message}")
```

```python
notas_df['aluno_mensagens'] = aluno_mensagens
```

```python
notas_df.to_csv("dados_msg.csv", index = False)
```
